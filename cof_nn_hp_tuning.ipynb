```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

from sklearn import metrics

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras import callbacks
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.utils import normalize
from keras.layers.advanced_activations import LeakyReLU
```

```python
features = pd.read_csv("gdrive/My Drive/cofs/combined_features.csv")
properties = pd.read_csv('gdrive/My Drive/cofs/properties.csv')
```

```python
# composition and structure to predict CO2Qst using NN

x = features
y = properties["CO2Qst_kJ_mol_"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 42)

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

model = Sequential()
model.add(Dense(372, activation = LeakyReLU(alpha=0.3)))
model.add(Dense(1000, activation = LeakyReLU(alpha=0.3)))
model.add(Dense(1, activation = LeakyReLU(alpha=0.3)))

model.compile(loss='mean_squared_error', optimizer='adam')

history = model.fit(x_train_scaled, y_train, epochs = 100)
```

```python
model.evaluate(x_test_scaled, y_test)

y_pred = model.predict(x_test_scaled)
print('R^2:', metrics.r2_score(y_test, y_pred))
```

```python
# hyper parameter tuning NN

def build_model(n_hidden = 1, n_neurons = 30, learning_rate = 3e-3, input_shape = 372):
  model = keras.models.Sequential()
  Dense(372, activation = LeakyReLU(alpha=0.3))
#  model.add(keras.layers.InputLayer(input_shape = input_shape))
  for layer in range(n_hidden):
    model.add(keras.layers.Dense(n_neurons, activation = LeakyReLU(alpha=0.3)))
    model.add(keras.layers.Dense(1))
    optimizer = keras.optimizers.Adam(learning_rate = learning_rate)
    model.compile(loss = "mse", optimizer = optimizer)
    return model

regressor = keras.wrappers.scikit_learn.KerasRegressor(build_model)
```

```python
# hyperparameter tuning for structural information from chemical

x = features.drop(['density_kg_m_3_', 'poreVolume_cm_3_g_',
       'largestIncludedSphereDiameter_A_', 'largestFreeSphereDiameter_A_',
       'largestIncludedSphereAlongFreeSpherePathDiameter_A_',
       'voidFraction_widom_'], axis = 1)
y_dens = features["density_kg_m_3_"]

x_train, x_test, y_train, y_test = train_test_split(x, y_dens, test_size = 0.25, random_state = 42)

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

params = {
    "n_hidden": [1,2,3],
    "n_neurons": np.arange(1,1000,100),
    "learning_rate": [3e-4],
}

rnd_src = RandomizedSearchCV(
    regressor,
    param_distributions = params,
    n_iter = 10,
)

rnd_src.fit(
    x_train,
    y_train,
    epochs = 100,
    validation_split = 0.2,
    callbacks = [keras.callbacks.EarlyStopping(patience=10)]
)
```
